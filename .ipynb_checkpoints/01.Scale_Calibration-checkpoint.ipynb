```python
#lmfit provides curve fiting using non-linear least squares minimization
#lmfit was developed by Matt Newville: https://github.com/newville

#install lmfit
from subprocess import call
call(['pip', 'install', 'lmfit'])
call(['pip', 'install', 'matplotlib'])

#import libraries
from numpy import sqrt, pi, exp, linspace, loadtxt
from lmfit.models import SkewedGaussianModel, GaussianModel, LorentzianModel, VoigtModel, PseudoVoigtModel, LinearModel, ConstantModel
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np

%matplotlib inline 
```

```python
#import the data
#The data was cleaned and prepared using external spreadsheet applications
#The data includes estimated errors based on instruments' accuracy
data = loadtxt('./Data/xyFP_and_yerr.dat')
x = data[:, 0]
y = data[:, 1]
yerr = data[:, 2]
```

```python
#define aggregate model using builtin models
#specify an alias(prefix) for each model:
mod = LinearModel(prefix='l0_') + \
      SkewedGaussianModel(prefix='g1_') + \
      LinearModel(prefix='l1_') + \
      SkewedGaussianModel(prefix='g2_') + \
      LinearModel(prefix='l2_')
```

```python
#initialize the parameters to be optimized
pars  = mod.make_params(l0_m=0.0, l0_b=0.0,
                        g1_amplitude=3.733, g1_center=0.016, g1_sigma=0.001698, g1_gamma=359.0957956, g1_height=3.57,
                        l1_m=0.0, l1_b=0.0,
                        g2_amplitude=3.268, g2_center=0.111, g2_sigma=0.00085, g2_gamma=955818.846, g2_height=3.268,
                        l2_m=6.99, l2_b=-0.624)
```

```python
# fit with weights=1/yerr
result = mod.fit(y, pars, x=x, weights=1./yerr)


#print the Fit Statistics and parameters
print(result.fit_report())
```

```python
# extract Model components
comps = result.eval_components(result.params, x=x)
```

```python
#define a dimensional annotation function
def annotate_dim(ax,xyfrom,xyto,text=None):

    if text is None:
        text = str()

    ax.annotate("",xyfrom,xyto,arrowprops=dict(arrowstyle='<->'))
    ax.text((xyto[0]+xyfrom[0])/2.0,(((xyto[1]+xyfrom[1])/2.0) + 0.03),text,fontsize=12)
```

```python
#plot data, model and error

# Plot the data points with error bars in yellow ('y')
plt.errorbar(x, y, yerr=yerr,fmt='none', ecolor='y', linewidth=0.5 )

# Plot the data points in blue ('b') 
plt.plot(x, y, 'b-', linewidth=3)
# Plot the model (result2.best_fit) in red ('r')
plt.plot(x, result.best_fit, 'r-', linewidth=3)

#Customize the graph and make annotations
plt.ylabel('Intensity [V]')
plt.xlabel('Time [s]')
plt.title('Resonance Peaks')

#Add vertical lines to mark the position of the centers of the resonance peaks
pl.axvspan(0.01599929, 0.11099999, edgecolor='green', facecolor='None', alpha=0.5)

#Plot the legend
plt.legend(('Data', 'Model'), shadow=True, loc=('upper right'), handlelength=0.7, fontsize=10)

#Annotate dimensions
annotate_dim(plt.gca(),[0.01599929, 3.5],[0.11099999, 3.5],'$0.095s$')

#Add info about fit coefficients and data
plt.figtext(0.4, 0.35, 'Data Points $= 1401$ \n$\chi^{2} =  1389.2$ \n$\chi^{2}_{red} = 1.002$', color='gray')  


#export the graph to image file (uncomment the next line to export the graph to an image file)
#plt.savefig('calibration_peaks.png', dpi=300)

plt.show()
```
