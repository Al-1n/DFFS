```python
#Import libraries

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
# Load data from files

#The data was cleaned and prepared using external spreadsheet applications
#The data includes estimated errors based on instruments' accuracy

data = np.loadtxt('./Data/xyMHz_and_yerr.dat')
x = data[:, 0]
y = data[:, 1]
yerr = data[:, 2]

data2 = np.loadtxt('./Data/xyMdf_and_yerr.dat')
y2 = data2[:, 1]
```

```python
# Define a Gaussian function to fit
def gaussian(x, amplitude, center, sigma):
    return amplitude * np.exp(-(x - center)**2 / (2 * sigma**2))
```

```python
# Define a linear function for the background
def linear(x, m, b):
    return m * x + b
```

```python
# Initial guess for parameters
initial_guess = [0.13, 800.0, 266.8, 0.0, 0.017]
```

```python
# Perform the fit using curve_fit
params, covariance = curve_fit(
    lambda x, amplitude, center, sigma, m, b: gaussian(x, amplitude, center, sigma) + linear(x, m, b),
    x, y, p0=initial_guess, sigma=yerr
)
```

```python
# Extract the fitted parameters
g1_amplitude, g1_center, g1_sigma, l1_m, l1_b = params

# Calculate the FWHM from the sigma of the Gaussian
fwhm = 2 * np.sqrt(2 * np.log(2)) * g1_sigma

# Calculate the fitted curve
fit_curve = gaussian(x, g1_amplitude, g1_center, g1_sigma) + linear(x, l1_m, l1_b)

# Calculate the chi-square and reduced chi-square
residuals = y - fit_curve
chi_square = np.sum((residuals / yerr)**2)
num_params = len(params)
num_data_points = len(x)
reduced_chi_square = chi_square / (num_data_points - num_params)
```

```python
# Print fit results
print("Amplitude:", g1_amplitude)
print("Center:", g1_center)
print("Sigma:", g1_sigma)
print("FWHM:", fwhm)
print("Linear Slope (m):", l1_m)
print("Linear Intercept (b):", l1_b)
print("Chi-Square:", chi_square)
print("Reduced Chi-Square:", reduced_chi_square)
```

```python
# Plot the data, the fitted curve, and the Doppler-free signal

# Create a new figure 
plt.figure(figsize=(9, 7))

plt.errorbar(x, y, yerr=yerr, fmt='none', ecolor='y', linewidth=0.2)
plt.plot(x, y, '#0099ff', linewidth=2, label='Doppler-broadened signal')
plt.plot(x, fit_curve, 'r-', linewidth=3, label='Model')
plt.plot(x, y2, '#996699', linewidth=2, label='Doppler-free signal')

# Set labels and add a green shaded region
plt.ylabel('Absorption [arb. units]')
plt.xlabel('Relative Frequency [MHz]')
plt.axvspan(479.256483, 1107.486207, facecolor='g', alpha=0.5)

# Add text annotations
plt.figtext(0.142, 0.7, 'Data Points $= 3550$ \n$\chi^{2} =  3545.027$ \n$\chi^{2}_{red} = 1.000$', color='#4a4a4a')
plt.figtext(0.44, 0.2, f'$FWHM = {fwhm:.4f} MHz$')

# Add legend
plt.legend(loc='lower left', bbox_to_anchor=(0.1, 1.0), ncol=3, borderaxespad=0, frameon=False)

# Show the plot
plt.show()
```
